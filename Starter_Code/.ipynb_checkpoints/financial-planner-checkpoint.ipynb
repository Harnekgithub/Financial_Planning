## Part 1 - Personal Finance Planner

### Collect Crypto Prices Using the `requests` Library

```python
# Set current amount of crypto assets
# set weightage for each coin
my_btc = 1.2
my_eth = 5.3
```

```python
# Crypto API URLs
#set up the paths(URLs) to access cyrpto data for Bitcoin and Etherium
btc_url = "https://api.alternative.me/v2/ticker/Bitcoin/?convert=CAD"
eth_url = "https://api.alternative.me/v2/ticker/Ethereum/?convert=CAD"
```

```python
# Fetch current BTC price
# used request.get() function to retrieve BTC data from the source
btc_data = requests.get(btc_url).json()

# Fetch current ETH price
# used request.get() function to retrieve ETHE data from the source
eth_data = requests.get(eth_url).json()

# Compute current value of my crpto
#use respective codes from the dataset to retrieve respective closing prices
my_btc_value = my_btc * btc_data['data']["1"]['quotes']['USD']['price']
my_eth_value = my_eth * eth_data['data']["1027"]['quotes']['USD']['price']

# Print current crypto wallet balance
print(f"The current value of your {my_btc} BTC is ${my_btc_value:0.2f}")
print(f"The current value of your {my_eth} ETH is ${my_eth_value:0.2f}")
```

```python
# Get the closing prices from that data frame
# Compute the current value of shares
agg_close_price = float(closing_price_df['AGG']['close'])
spy_close_price = float(closing_price_df['SPY']['close'])
my_agg_value = my_agg * agg_close_price
my_spy_value = my_spy * spy_close_price

# Print current value of shares
print(f"The current value of your {my_spy} SPY shares is ${my_spy_value:0.2f}")
print(f"The current value of your {my_agg} AGG shares is ${my_agg_value:0.2f}")
```

### Savings Health Analysis

```python
# Setting the monthly household income to $1200
monthly_income = 1200

# Add all the crypto and stock amounts 

crypto_amt = my_btc_value + my_eth_value
shares_amt = my_agg_value + my_spy_value

# Setup the dictionary to create the savings df
data = {'Amount':[crypto_amt, shares_amt]}

# Create savings DataFrame using pandas DataFrame function
df_savings = pd.DataFrame(data, index=['crypto','shares'])

# Validate that data frame was created with right data and structure
display(df_savings)
```

```python
# Visualize the savings dataframe using pie chart
pie_plot = df_savings.plot.pie(y="Amount", title="Stocks Portfolio Composition")
```

```python
# Set ideal emergency fund to three times the monthly income
emergency_fund = monthly_income * 3

# Calculate total amount of savings by adding crypto and stock values
total_savings = crypto_amt + shares_amt

# Elavuate if the saving are enough or less than the emergency fund
if total_savings > emergency_fund:
    print(f"Congratulations your Total savings {total_savings} are more than the emergency fund needed 3x the monthly income {emergency_fund}")
elif total_savings == emergency_fund:
    print(f"Congratulations your Total savings {total_savings} mmet the emergency fund needed that is 3x the monthly income {emergency_fund}")
else:
    print(f"your Total savings {total_savings} are {emergency_fund - total_savings} less than the emergency fund needed 3x the monthly income {emergency_fund}")
```

## Part 2 - Retirement Planning

### Monte Carlo Simulation

```python
# Set start and end dates of five years back from today.
start_date = pd.Timestamp('2016-09-17', tz='America/New_York').isoformat()
end_date = pd.Timestamp('2021-09-16', tz='America/New_York').isoformat()
```

```python
# Get 5 years' worth of historical data for SPY and AGG
# (use a limit=1000 parameter to call the most recent 1000 days of data)
#setup weightage and number of simulations to be done
weights = [.40, .60]
num_simulations = 500

#initialize the number of year to 30years to run simulation 
num_days = 252*30

#get the data by calling the get_barset function of the alpaca object to retrieve the data
ticker_data = alpaca_api_df.get_barset(tickers,timeframe, start = start_date , end = end_date, limit=1000 ).df

# check for data retrieved 
ticker_data.head()
```

```python
# Configuring a Monte Carlo simulation to forecast 30 years cumulative returns
mc_data  = MCSimulation(ticker_data, weights, num_simulations, num_days)
```

```python
# Print the simulation input data
mc_data.portfolio_data
```

```python
# Run a Monte Carlo simulation to forecast 30 years cumulative returns
mc_30year = mc_data.calc_cumulative_return()
```

```python
# Plot simulation outcomes
simu_plot = mc_data.plot_simulation() 
```

```python
# Plot probability distribution and confidence intervals
dist_plot = mc_data.plot_distribution()
```

### Retirement Analysis

```python
# Fetch summary statistics from the Monte Carlo simulation results
sum_data = mc_data.summarize_cumulative_return()

# Print summary statistics
sum_data
```

### Calculate the expected portfolio return at the `95%` lower and upper confidence intervals based on a `$20,000` initial investment.

```python
# Set initial investment to $20000
initial_investment = 20000

# Use the lower and upper `95%` confidence intervals to calculate the range of the possible outcomes of our $20,000
ci_lower = round(sum_data[8] * initial_investment, 2)
ci_upper = round(sum_data[9] * initial_investment, 2)

# Print results
print(f"There is a 95% chance that an initial investment of ${initial_investment} in the portfolio"
      f" over the next 30 years will end within in the range of"
      f" ${ci_lower} and ${ci_upper}")
```

### Calculate the expected portfolio return at the `95%` lower and upper confidence intervals based on a `50%` increase in the initial investment.

```python
# Set initial investment by increaseing it with 50% more initial investment
initial_investment = 20000 * 1.5

# Use the lower and upper `95%` confidence intervals to calculate the range of the possible outcomes of our $30,000
ci_lower_30k = round(sum_data[8] * initial_investment, 2)
ci_upper_30k = round(sum_data[9] * initial_investment, 2)

# Print results
print(f"There is a 95% chance that an initial investment of ${initial_investment} in the portfolio"
      f" over the next 30 years will end within in the range of"
      f" ${ci_lower_30k} and ${ci_upper_30k}")
```
